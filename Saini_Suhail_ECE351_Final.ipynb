```python
#Question 2

def x1(t):
    return np.cos(-3*np.pi*t)

def x2(t):
    return 6*np.sin(1.6*np.pi*t)

def x3(t):
    return 2*np.e**(-0.5*t)*2*np.sin(3.7*np.pi*t)

def x4(t):
    return np.e**(-6*t) + 5*np.e**(-3*t) + 10*np.e**(-t)

step = 1e-3
t = np.arange(0, 8, step)

plt.figure(figsize = (12,12))
plt.suptitle('ECE 351 Final - Question 2')

plt.subplot(2,2,1)
plt.plot(t, x1(t), 'k', label = 'cos(-3$\pi$t)')
plt.grid()
plt.title('$x_1$')
plt.ylabel('x(t)')
plt.xlabel('t[s]')
plt.ylim(-1.5, 1.5)
plt.xlim(0, 8)
plt.locator_params(axis = "both", integer = True, tight = True)
plt.legend(loc ='upper right')

plt.subplot(2,2,2)
plt.plot(t, x2(t), 'b', label = '6sin(1.6$\pi$t)')
plt.grid()
plt.title('$x_2$')
plt.ylabel('x(t)')
plt.xlabel('t[s]')
plt.ylim(-8, 8)
plt.xlim(0, 8)
plt.locator_params(axis = "both", integer = True, tight = True)
plt.legend(loc ='lower left')

plt.subplot(2,2,3)
plt.plot(t, x3(t), 'm', label = '2$e^{-0.5t}$2sin(3.7$\pi$t)')
plt.grid()
plt.title('$x_3$')
plt.ylabel('x(t)')
plt.xlabel('t[s]')
plt.ylim(-4, 4)
plt.xlim(0, 8)
plt.locator_params(axis = "both", integer = True, tight = True)
plt.legend(loc ='upper right')

plt.subplot(2,2,4)
plt.plot(t, x4(t), 'g', label = '$e^{-6t}$+$5e^{-3t}$+$10e^{-t}$')
plt.grid()
plt.title('$x_4$')
plt.ylabel('x(t)')
plt.xlabel('t[s]')
plt.ylim(-1, 16)
plt.xlim(0, 8)
plt.locator_params(axis = "both", integer = True, tight = True)
plt.legend(loc ='upper right')

plt.tight_layout()
plt.show()
```
